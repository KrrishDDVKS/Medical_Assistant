```python
def summarize_transcipt(transcript:str) -> str:
    chat_completion = client.chat.completions.create(
        messages=[
            {"role": "system", "content": "Read the following Patient-Doctor transcript in a cardiovascular-disease-clinic and summarize it. \
            Retain all technimal medical terms and focus on patient symptoms and Doctor's response."},
            {"role": "user", "content": transcript}
        ],
        model="gpt-4o-mini",
        max_tokens=150
    )

    return chat_completion.choices[0].message.content
```

```python
folder_path = '/content/drive/MyDrive/Clean Transcripts/'

# Initialize an empty dictionary to store summaries
summaries = {}

# Iterate through each file in the specified folder
for filename in os.listdir(folder_path):
    # Check if the file is a .txt file
    if filename.endswith('.txt'):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        print(filename)
        # Open the file and read its contents into a string
        with open(file_path, 'r') as file:
            transcript = file.read()

        # Call the pre-defined function and store the result
        summary = summarize_transcipt(transcript)

        print(summary)
        print()
        # Store the summary in the dictionary using the file name as the key
        filename=filename.split('.')[0]
        summaries[filename] = summary
```

```python
import pandas as pd
df=pd.DataFrame(summaries.items(), columns=['File Name', 'Summary'])
df.to_csv('summaries.csv', index='File Name')
df
```
